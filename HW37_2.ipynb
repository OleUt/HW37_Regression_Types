```python
import pandas as pd
data = pd.read_csv('communities.csv')
data
```

```python
# drop columns with missing data, drop duplicates

df = data.replace('?', None).dropna(axis='columns', how='any')  #, inplace=True)
df.drop_duplicates()
df
```

```python
# columns names

df.columns = [f'C{i}' for i in range(1, 104)]
df
```

```python
cols = list(df.columns)
a, b = cols.index('C1'), cols.index('C2')
cols[a], cols[b] = cols[b], cols[a]
df = df[cols]
df
```
